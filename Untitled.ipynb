```python
conditions=[
   
    ((dff['PROPERTY_TYPE']=='House')|(dff['PROPERTY_TYPE']=='Bungalow')) & 
    ((((((dff['TRANSACTION_TYPE']=='rental (private)')|(dff['TRANSACTION_TYPE']=='FiT application')|(dff['TRANSACTION_TYPE']=='RHI application')|(dff['TRANSACTION_TYPE']=='assessment for green deal')|
     (dff['TRANSACTION_TYPE']=='new dwelling')| (dff['TRANSACTION_TYPE']=='following green deal')))))) &
    (((dff['BUILT_FORM']=='Semi-Detached')|(dff['BUILT_FORM']=='Detached')|(dff['BUILT_FORM']=='End-Terrace')))&
    ((((dff['CURRENT_ENERGY_RATING']=='A')|(dff['CURRENT_ENERGY_RATING']=='B')|(dff['CURRENT_ENERGY_RATING']=='C')|(dff['CURRENT_ENERGY_RATING']=='D'))))&
    (((dff['ROOF_DESCRIPTION']!='Flat')|(dff['ROOF_DESCRIPTION']!='Thatched')|(dff['ROOF_DESCRIPTION']!='0')))
    & (dff['LODGEMENT_DATE']>'01-04-2012')
    
]
#choices=['high','medium','average','average']
choices=['high']
dff['solar_suitable']=np.select(conditions,choices,default='low')
```
